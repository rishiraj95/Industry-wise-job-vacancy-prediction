### There are two types of industry codes in use, the NAICS code and QCEW higher level code. The ones which start with 10, could not be further classified into more specific categories and hence should go into others. 

```python
naics_code_lookup = {'11':'Agriculture, Forestry, Fishing, Hunting', '21': 'Mining', '22': 'Utilities', 
                     '23': 'Construction', '31': 'Manufacturing', '32': 'Manufacturing', '33': 'Manufacturing', 
                     '42':'Wholesale trade', '44': 'Retail trade', '45': 'Retail trade', 
                     '48': 'Transportation, Warehousing', '49': 'Transportation, Warehousing', 
                     '51': 'Information', '52': 'Finance, Insurance', '53':'Real Estate', 
                     '54': 'Professional, Scientific, Technical services', 
                     '55': 'Management of companies, enterprises', 
                     '56': 'Waste management, remediation, administrative services', 
                     '61': 'Educational services', '62': 'Health care, social assistance', 
                     '71': 'Arts, Entertainment, Recreation', '72': 'Accomodation, Food services', 
                     '81': 'Other services', '92': 'Public administration'}

qcew_code_lookup = {'10': 'Other services', '101': 'Manufacturing', '1011': 'Mining', '1012': 'Construction', '1013': 
                    'Manufacturing', '102': 'Other services', '1021': 'Transportation, Warehousing', '1022': 'Information',
                   '1023': 'Finance, Insurance', '1024': 'Professional, Scientific, Technical services',
                   '1025': 'Health care, social assistance', '1026': 'Arts, Entertainment, Recreation',
                   '1027': 'Other services', '1028': 'Public administration', '1029': 'Other services'}
def shrink_indus_cat(row):
    return qcew_code_lookup[row['industry_code']] if row['industry_code'][:2] == '10' else naics_code_lookup[row['industry_code'][:2]]
```

```python
# Create a new column called industry_sector, merging alike industries using the industry code
Alabama_file['industry_sector'] = Alabama_file.apply(shrink_indus_cat,axis=1)
```

```python
Alabama_file.index.name='index'
```

```python
Alabama_file.groupby(['industry_sector']).count()[['area_fips']].sort_values(['area_fips'], ascending=False)
```
